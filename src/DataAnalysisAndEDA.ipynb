```python
import pandas as pd
import numpy as np
import seaborn as sns
from IPython.display import display
import glob
from matplotlib import pyplot
pal = sns.set(style="whitegrid", color_codes=True)
```

```python
file_pths = glob.glob("/content/drive/MyDrive/UberData/*.csv")

data = {}

for pth in file_pths:
    print(pth)
    df = pd.read_csv(pth)
    display(df)

    data[pth] = df
```

```python
month_total = np.zeros(11)
month_perc = np.zeros(11)

unemployment = data["/content/drive/MyDrive/UberData/unemployment.csv"]
population = data["/content/drive/MyDrive/UberData/population.csv"]
destination_data2 = data["/content/drive/MyDrive/UberData/immigrants_emigrants_by_destination2.csv"]
destination_data1 = data["/content/drive/MyDrive/UberData/immigrants_emigrants_by_destination.csv"]

pop_sum2 = population[population["Year"]==2017].groupby("District.Name")["Number"].sum()

for month in unemployment["Month"].unique():
    pop_sum1 = unemployment[unemployment["Year"]==2017][unemployment["Month"]==month].groupby("District Name")["Number"].sum()

    total_employed = pop_sum2 - pop_sum1
    perc_employed = total_employed / pop_sum2 * 100

    month_total += total_employed.values
    month_perc += perc_employed.values

month_total /= 12
month_perc /= 12
```

```python
import math
total = 0
for value in month_total:
    if not math.isnan(value):
        total += value

df = pd.DataFrame({"District Name" : pop_sum2.keys(),
                   "Percentage Employed" : [x for i, x in enumerate(month_perc) if i != 5],
                   "Population Size" : pop_sum2.values,
                   "Total Employed" : [x for i, x in enumerate(month_total) if i != 5]})
```

```python
def get_num_traveled(district_name):
    immigrants_moved_from = destination_data2[destination_data2["from"] == district_name]["weight"].sum()
    immigrants_moved_to = destination_data2[destination_data2["to"] == district_name]["weight"].sum()

    return immigrants_moved_from + immigrants_moved_to

df["Number Traveled"] = df["District Name"].apply(get_num_traveled)
df["Population Weightage"] = df["Population Size"] = df["Population Size"] / df["Population Size"].max()
df["Travel Coefficient"] = df["Number Traveled"] / df["Total Employed"] * 100 * df["Population Weightage"]
```

```python
import matplotlib.pyplot as plt
sns.set(font_scale=2)
pal = sns.color_palette("viridis", 10)
fig, ax = pyplot.subplots(figsize=(20, 10))
sns.barplot(y=df["Travel Coefficient"], x=df["District Name"], palette=pal)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.savefig('Travel_Relative_To_Employment_and_Population.png')
```

```python
new_df = pd.read_csv("/content/drive/MyDrive/UberData/exiample_hotels_NEW")

vals1 = new_df["cumulativeTime"].values
vals2 = new_df["numDrivingBetter"].values

def scale_distribution(x, a, b, min1, max1):
    return ((b - a) * (x - min1)) / (max1 - min1) + a
```

```python
cum_max = max(vals1)
cum_min = min(vals1)

dri_max = max(vals2)
dri_min = min(vals2)

freq_max = max(freq_counts)
freq_min = min(freq_counts)

new_vals1 = []
new_vals2 = []

for i, val in enumerate(vals1):
    if geopy.distance.geodesic([2.1650, 41.3923], exiample_hotel_data.iloc[i][["Longitude", "Latitude"]].values).km * 0.621371 < 1.3:
        new_val = scale_distribution(val, freq_min, freq_max, cum_min, cum_max)
        new_vals1.append(new_val)

for i, val in enumerate(vals2):
    if geopy.distance.geodesic([2.1650, 41.3923], exiample_hotel_data.iloc[i][["Longitude", "Latitude"]].values).km * 0.621371 < 1.3:
        new_val = scale_distribution(val, freq_min, freq_max, dri_min, dri_max)
        new_vals2.append(new_val)

final_values = freq_counts + (1 - np.array(new_vals1)) + (1 - np.array(new_vals2))
```

```python
results = np.zeros(len(final_values))

for v in np.argsort(final_values)[:5]:
    results[v] = 1
```
